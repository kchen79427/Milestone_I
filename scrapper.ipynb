```python

# json_scraper
# reads the first race in daytona and increments to the next race , every time it is call.
def json_scraper():
    def json_locator(url):
        import json
        import re

        base_url = url

        pattern= r'\d{4}(?=/lap-times)'
        new_race=re.findall(pattern,base_url)
        new_race=int(new_race[0])+1

        data_to_grab=rf'https://cf.nascar.com/cacher/2022/1/{new_race}/lap-times.json'
        return data_to_grab, new_race

    race_id_json={
    "Daytona International Speedway":5146,
    "Auto Club Speedway":5147,
    "Las Vegas Motor Speedway":5148,
    "Phoenix Raceway":5149,
    "Atlanta Motor Speedway":5150,
    "Circuit of the Americas":5151,
    "Richmond Raceway":5152,
    "Martinsville Speedway":5153,
    "Bristol Motor Speedway (dirt)":5154,
    "Talladega Superspeedway":5155,
    "Dover Motor Speedway":5156,
    "Darlington Raceway":5157,
    "Kansas Speedway":5158,
    "Texas Motor Speedway":5159,
    "Texas Motor Speedway_II":5160,
    "Charlotte Motor Speedway":5161,
    "World Wide Technology Raceway":5162,
    "Sonoma Raceway":5163,
    "Nashville Superspeedway":5164,
    "Road America":5165,
    "Atlanta Motor Speedway_II":5166,
    "New Hampshire Motor Speedway":5167,
    "Pocono Raceway":5168,
    "Indianapolis Motor Speedway Road Course":5169,
    "Michigan International Speedway":5170,
    "Richmond Raceway_II":5171,
    "Watkins Glen International":5172,
    "Daytona International Speedway_II":5173,
    "Darlington Raceway_II":5174,
    "Kansas Speedway_II":5175,
    "Bristol Motor Speedway":5176,
    "Texas Motor Speedway_III":5177,
    "Talladega Superspeedway_II":5178,
    "Charlotte Motor Speedway ROVAL":5179,
    "Las Vegas Motor Speedway_II":5180,
    "Homestead-Miami Speedway":5181,
    "Martinsville Raceway":5182,
    "Phoenix Raceway_II":5183}

    def get_key(val):
        for key, value in race_id_json.items():
            if val == value:
                return key
    
        return "key doesn't exist"

    with open(r'data\json_url.txt','r+') as json_txt:
        # check this file for the next url
        unpack_me=json_txt.readlines()
        updated_url=json_locator(unpack_me[0])
        update=updated_url[0]
        race_id=int(updated_url[1])
        json_txt.seek(0)
        json_txt.write(update)
        json_txt.write('\n')
        json_txt.write(get_key(race_id))
        json_txt.truncate()
        
        import requests, json
        data=requests.get(update)
        data=json.loads(data.text)
        
    return data
```
